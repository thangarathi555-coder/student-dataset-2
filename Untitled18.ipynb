```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(101)   # different seed (not same as yours)

n = 250
```

```python
student_ids = np.arange(1, n+1)
names = np.array(["Student_" + str(i) for i in student_ids])

genders = np.random.choice(["Male", "Female"], n, p=[0.55, 0.45])
departments = np.random.choice(["CSE", "IT", "ECE"], n)
years = np.random.choice(["1st", "2nd", "3rd"], n)
```

```python
maths = np.clip(np.random.normal(70, 15, n), 35, 100).astype(int)
science = np.clip(np.random.normal(68, 14, n), 35, 100).astype(int)
english = np.clip(np.random.normal(72, 12, n), 35, 100).astype(int)
```

```python
attendance = np.clip(np.random.normal(80, 10, n), 50, 100).astype(int)
```

```python
internal_marks = (maths + science + english) // 15
internal_marks = np.clip(internal_marks, 0, 25)
```

```python
df = pd.DataFrame({
    "Student_ID": student_ids,
    "Name": names,
    "Gender": genders,
    "Department": departments,
    "Year": years,
    "Maths": maths,
    "Science": science,
    "English": english,
    "Attendance": attendance,
    "Internal_Marks": internal_marks
})
```

```python
df["Total"] = df["Maths"] + df["Science"] + df["English"] + df["Internal_Marks"]
df["Average"] = df["Total"] / 4
```

```python
def grade(avg):
    if avg >= 85:
        return "A"
    elif avg >= 70:
        return "B"
    elif avg >= 50:
        return "C"
    else:
        return "F"

df["Grade"] = df["Average"].apply(grade)
```

```python
df.to_csv("student_dataset_realistic.csv", index=False)

print(df.head())
```

```python
df.groupby("Department")["Average"].mean().plot(kind="bar")
plt.title("Department-wise Performance")
plt.show()
```

```python
df["Gender"].value_counts().plot(kind="pie", autopct='%1.1f%%')
plt.show()
```

```python
plt.hist(df["Average"])
plt.title("Average Marks Distribution")
plt.show()
```
